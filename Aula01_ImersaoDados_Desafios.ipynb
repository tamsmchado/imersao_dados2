## Bioinformática

O que é Bioinformática e qual sua relação com a Ciência de Dados?

Segundo o site de pesquisa Wikipédia, Bioinformática é:

"Bioinformática é um campo interdisciplinar que corresponde à aplicação das técnicas da informática, no sentido de análise da informação, nas áreas de estudo da biologia. Como um campo interdisciplinar da ciência, a bioinformática combina a biologia, ciência da computação, estatística, matemática e engenharia para analisar, interpretar e processar dados biológicos."

Ainda lendo sobre Bioinformática na Wikipédia temos a seguente afirmação que nos apresenta uma relação direta entre Bioinformática e Ciência de Dados:

Os algoritmos, por sua vez dependem de fundamentos teóricos, tais como matemática discreta, teoria de controle, teoria de sistemas, teoria da informação e estatística."

## Droga

"Droga é toda e qualquer substância, natural ou sintética que, uma vez introduzida no organismo, modifica suas funções.

Em medicina, refere-se a qualquer substância com o potencial de prevenir ou curar doenças ou aumentar o bem-estar físico ou mental; em farmacologia, refere-se a qualquer agente químico que altera os processos bioquímicos e fisiológicos de tecidos ou organismos."

Link para mais informações: https://pt.wikipedia.org/wiki/Droga

## Grupo Controle (com_controle) x Grupo Experimental ou Tratamento (com_droga)

"Em um experimento , os dados de um grupo experimental são comparados com os dados de um grupo de controle. Esses dois grupos devem ser idênticos em todos os aspectos, exceto um: a diferença entre um grupo de controle e um grupo experimental é que a variável independente é alterada para o grupo experimental, mas é mantida constante no grupo de controle."

"Um grupo experimental é uma amostra de teste ou o grupo que recebe um procedimento experimental. Este grupo está exposto a mudanças na variável independente que está sendo testada. Os valores da variável independente e o impacto na variável dependente são registrados. Um experimento pode incluir vários grupos experimentais ao mesmo tempo."

"Um grupo de controle é um grupo separado do resto do experimento de forma que a variável independente sendo testada não possa influenciar os resultados. Isso isola os efeitos da variável independente no experimento e pode ajudar a descartar explicações alternativas dos resultados experimentais."



**Um bom exemplo pode facilitar muito:**

"Um exemplo simples de um experimento controlado pode ser usado para determinar se as plantas precisam ou não de ser regadas para viver. O grupo de controle seria composto por plantas que não são regadas. O grupo experimental seria composto por plantas que recebem água."

### Desafio 01: Ordenar as barras do gráfico countplot do maior para o menor.

O gráfico Countplot da bibiloteca Seaborn consiste em plotar a contagem dos elementos de acordo com o parâmetro que lhe for passado. Para este desafio foi proposto plotar os 5 compostos mais significativos da coluna "droga" ordenando os compostos por sua quantidade do maior para o menor, mas para realizar esse desafio precisamos realizar algumas mudanças na nossa base de dados. Como foi sugerido na aula, deve-se alterar o nome da coluna "droga" para "composto", isso faz sentido porque os elementos contidos na coluna "droga" ainda não são de fato drogas, são compostos que estão sendo testados com potencial de se tornarem drogas.
Então vamos precisar realizar alguns ajustes e segue abaixo o passo a passo.


1. Renomeando a coluna "droga" para "composto"

```python
dados.rename(columns={'droga': 'composto'}, inplace = True)
dados.head()
```

2. Criando a variável compostos que receberá a contagem dos elementos da coluna "composto" e o filtro dos 5 mais significativos.

3. Guardando na variável cod_composto o index do resultado gerado pela variável compostos, esse index represenda o nome dos top 5 compostos

```python
cod_compostos = dados['composto'].value_counts().index[:5]
cod_compostos
```

4. Fazendo um filtro na coluna "compostos" da base de dados utilizando o método query, na qual retornará a base de dados apenas com os elementos mais significativos

```python
dados.query('composto in @cod_compostos')
```

Essa "nova" base de dados filtrada será utilizada para criação do gráfico countplot, sendo passada no parâmetro "data".

###Desafio 04: Estudar sobre as estatísticas método describe()

Máximo e Mínimo
* "A mínima e a máxima são os valores nos “extremos” de um conjunto de dados ordenados: ou seja, o menor e o maior valor. Ambos são utilizados cotidianamente e facilmente compreensíveis para quase toda população, pense – por exemplo – em seu uso diário na previsão do tempo."



Média
* "Média (M), mais precisamente chamada de média aritmética simples, é o resultado da soma de todas as informações de um conjunto de dados dividida pelo número de informações que foram somadas."




Mediana
* "Se o conjunto de informações for numérico e estiver organizado em ordem crescente ou decrescente, a sua mediana será o número que ocupa a posição central da lista."



Desvio Padrão
* "Em probabilidade, o desvio padrão ou desvio padrão populacional é uma medida de dispersão em torno da média populacional de uma variável aleatória. O termo possui também uma acepção específica no campo da estatística, na qual também é chamado de desvio padrão amostral  e indica uma medida de dispersão dos dados em torno de média amostral. Um baixo desvio padrão indica que os pontos dos dados tendem a estar próximos da média ou do valor esperado. Um alto desvio padrão indica que os pontos dos dados estão espalhados por uma ampla gama de valores. O desvio padrão populacional ou amostral é a raiz quadrada da variância populacional ou amostral correspondente, de modo a ser uma medida de dispersão que seja um número não negativo e que use a mesma unidade de medida dos dados fornecidos."



Variação Interquartil (Q3-Q1)
* "Tanto o desvio padrão quanto a variância dependem da média, portanto, são afetados por valores extremos. Mas há uma medida de dispersão baseada na mediana, que consegue contornar a presença destes outliers: a variação interquartil, abreviada para IQR em inglês.
Para descobrir a variação interquartil, primeiramente, dividimos os dados em duas metades, como na mediana, para em seguida refazer esta operação com cada uma das duas metades. Obtemos então quatro partes iguais – ou quartis.
Agora, você subtrai a mediana da metade superior (terceiro quartil ou Q3) com a mediana da metade inferior (primeiro quartil ou Q1). A vantagem deste método é que ele, por definição, já isola os outliers, seja inferiores ou superiores."



OUTLIERS
* "São valores atípicos entre os dados analisados. Isto é avaliado a partir do quanto um determinado valor difere dos demais em um conjunto de dados. Eles podem trazer tanto uma má notícia, como erros na coleta ou limpeza dos dados, como também boas novas, como os casos que fogem da regra e podem ganhar destaque na sua análise.
Por um lado, os outliers podem revelar um erro a ser corrigido ou pelo menos considerado, antes de avançarmos com a análise. Por isso, é importante começar as análises identificando-os, pois eles podem ser indícios de erros de medição ou falhas na limpeza de dados."
