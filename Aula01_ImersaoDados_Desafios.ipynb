## Droga

"Droga é toda e qualquer substância, natural ou sintética que, uma vez introduzida no organismo, modifica suas funções.

Em medicina, refere-se a qualquer substância com o potencial de prevenir ou curar doenças ou aumentar o bem-estar físico ou mental; em farmacologia, refere-se a qualquer agente químico que altera os processos bioquímicos e fisiológicos de tecidos ou organismos."

Link para mais informações: https://pt.wikipedia.org/wiki/Droga

## Grupo Controle (com_controle) x Grupo Experimental ou Tratamento (com_droga)

"Em um experimento , os dados de um grupo experimental são comparados com os dados de um grupo de controle. Esses dois grupos devem ser idênticos em todos os aspectos, exceto um: a diferença entre um grupo de controle e um grupo experimental é que a variável independente é alterada para o grupo experimental, mas é mantida constante no grupo de controle."

"Um grupo experimental é uma amostra de teste ou o grupo que recebe um procedimento experimental. Este grupo está exposto a mudanças na variável independente que está sendo testada. Os valores da variável independente e o impacto na variável dependente são registrados. Um experimento pode incluir vários grupos experimentais ao mesmo tempo."

"Um grupo de controle é um grupo separado do resto do experimento de forma que a variável independente sendo testada não possa influenciar os resultados. Isso isola os efeitos da variável independente no experimento e pode ajudar a descartar explicações alternativas dos resultados experimentais."



**Um bom exemplo pode facilitar muito:**

"Um exemplo simples de um experimento controlado pode ser usado para determinar se as plantas precisam ou não de ser regadas para viver. O grupo de controle seria composto por plantas que não são regadas. O grupo experimental seria composto por plantas que recebem água."

## Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html

https://pandas.pydata.org/docs/reference/api/pandas.Series.unique.html

```python
dados['tratamento'].value_counts()
```

```python
dados['tratamento'].value_counts(normalize = True) 
```

```python
dados[dados['tratamento'] == 'com_controle']
```

Quando realizamos o filtro "com_controle" para a coluna "tratamento", verifica-se que a coluna "droga" aparenta receber apenas um tipo de droga. 
Para verificar se essa suposição é verdadeira aplica-se o método unique() para a coluna "droga" com os filtros "com_controle" e "com_droga" na coluna "tratamento", a fim de realizar a comparação valores recebidos na coluna "droga".

```python
dados['droga'][dados['tratamento'] == 'com_controle'].unique()
```

```python
dados['droga'][dados['tratamento'] == 'com_droga'].unique()
```

Suposição verdadeira!

Ao aplicar o método unique() vericamos que a coluna "tratamento" com valores "com_controle" só recebem um tipo de droga, 'cacb2b860'. Já os casos de "tratamento" e filtro "com_droga" recebem várias drogas diferenres. 
Isso ocorre, pois, assim como verificamos na seção "Grupo Controle (com_controle) x Grupo Experimental ou Tratamento (com_droga)", um experimeno pode incluir diversos grupos experimentais diferentes, neste caso, testando diferentes substâncias em cada grupo.

## Desafio 07: Deixar os gráficos bonitões(título, labels e etc). Usando matplotlib.pyplot

```python
dados['tratamento'].value_counts(normalize = True)
```

```python
import matplotlib.pyplot as plt

labels = dados['tratamento'].value_counts().index
plt.figure(figsize=(12,8))
ax = plt.subplot(1,2,1)
ax.pie(dados['tratamento'].value_counts(), labels = labels, autopct='%1.1f%%')
ax.set_title('Frequencia de Tratamento', fontsize = 18)

ax = plt.subplot(1,2,2)
ax.bar(x = dados['tratamento'].value_counts().index, height=dados['tratamento'].value_counts())
ax.set_xlabel('Tipo de Tratamento', fontsize = 12)
ax.set_ylabel('Quantidade de Elementos', fontsize = 12)
ax.set_title('Frequencia de Tratamento', fontsize = 18)

plt.tight_layout(pad=4)
plt.show()

```

```python
quantidade_elementos = dados['tempo'].value_counts()
tempo_exposicao =  dados['tempo'].unique()

plt.xticks(tempo_exposicao)
plt.xlabel('Tempo de Exposição em horas', fontsize = 12)
plt.ylabel('Quantidade de Elementos', fontsize = 12)
plt.title('Quantidade de Elementos por Tempo de Exposição', fontsize = 18)
plt.bar(tempo_exposicao, quantidade_elementos, width=15)
plt.show()
```
